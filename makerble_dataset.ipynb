#Inferential Statistics

#Hypothesis test

```python
from scipy import stats
```

Null Hypothesis (H0): There is no difference in average scores of Job satisfaction between IT and HR departments.


Alternative Hypothesis (H1): There is a difference in average scores of Job Satisfsction between IT and HR departments.

 Extract job satisfaction scores for each department

```python
df['Job Satisfaction']=df['Job Satisfaction'].replace({'Strongly Disagree':0,'Disagree':1,'Neutral':2,'Agree':3,'Strongly Agree':4})
it_scores = df[df['Department'] == 'IT']['Job Satisfaction']
hr_scores = df[df['Department'] == 'HR']['Job Satisfaction']
```

```python
print(it_scores)
print(hr_scores)
```

```python
t_stat, p_value = stats.ttest_ind(it_scores, hr_scores)
```

```python
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
alpha = 0.50

if p_value <= alpha:
    print("Reject the null hypothesis (H0): There is a significant difference in job satisfaction between IT and HR.")
else:
    print("Fail to reject the null hypothesis (H0): There is no significant difference in job satisfaction between IT and HR.")
```

#Chi-Square Test

```python
filtered_df = df[df['Department'].isin(['IT', 'HR'])]
```

```python
contingency_table = pd.crosstab(filtered_df['Department'], filtered_df['Job Satisfaction'])
print("Contingency Table:\n", contingency_table)
```

```python
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
```

```python
print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)
```

```python
alpha = 0.05
if p_value >= alpha:
    print("Reject the null hypothesis (H0): There is a significant difference in job satisfaction between IT and HR.")
else:
    print("Fail to reject the null hypothesis (H0): There is no significant difference in job satisfaction between IT and HR.")
```
